# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
#load and check marketing_customer_analysis.csv
marketing_customer_df = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')
marketing_customer_df
```

```python
#Checking info about marketing_customer_df
marketing_customer_df.info()
```

```python
#Check value_count for marketing_customer_df.columns
for column in marketing_customer_df.columns:
    print(f"Value counts for {column}:")
    print(marketing_customer_df[column].value_counts())
    print("\n")
```

```python
#Need to delete unnamed:_0 column
marketing_customer_df_clean = marketing_customer_df.drop(columns='unnamed:_0')
marketing_customer_df_clean
```

```python
#Column 'effective_to_date' should be in datetime format, yyyy-mm-dd

marketing_customer_df_clean['effective_to_date'] = pd.to_datetime(marketing_customer_df['effective_to_date'], errors='coerce')
#display(marketing_customer_df_clean.head())       

#print(marketing_customer_df.dtypes)
#print(marketing_customer_df_clean.dtypes)

#column 'months_since_last_claim' (col[14]) --> dtype change to int64

#marketing_customer_df['months_since_last_claim']= marketing_customer_df['months_since_last_claim'].applymap(lambda x:int(x) if isinstance (x, (float,int)) else x) 
marketing_customer_df_clean['months_since_last_claim'] = marketing_customer_df['months_since_last_claim'].round().astype('int64') 

#display(marketing_customer_df_clean['months_since_last_claim'].value_counts())
#display(type(marketing_customer_df['months_since_last_claim']))

#The current dtype is float64. This should be int64, not a float64 number as you can not have a half of a complaint.
marketing_customer_df_clean['number_of_open_complaints'] = marketing_customer_df['number_of_open_complaints'].round().astype('int64') 

#Change column name 'employmentstatus' with '_' for the consistency of column name
marketing_customer_df_clean=marketing_customer_df_clean.rename(columns = {'employmentstatus': 'employment_status'})

#Values in 'vehicle_class': 'Sport cars', 'luxury SUV', and 'luxury' could be replaced by 'luxury'
marketing_customer_df_clean['vehicle_class']=marketing_customer_df_clean['vehicle_class'].apply(lambda x: 'Luxury' if x in ['Sports Car', 'Luxury SUV', 'Luxury Car'] else x) 
#display(marketing_customer_df_clean['vehicle_class'].value_counts())

#Column vehicle_size: replace 'medsize' with 'Medium'
marketing_customer_df_clean['vehicle_size'] = marketing_customer_df_clean['vehicle_size'].apply(lambda x: 'Medium' if x == 'Medsize' else x )
display(marketing_customer_df_clean['vehicle_size'].value_counts())

#print(marketing_customer_df_clean.dtypes)
display(marketing_customer_df_clean.head())
```

Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
#Check this dataset is from one year or from multiple years
check = pd.DataFrame()
check['year'] = marketing_customer_df_clean['effective_to_date'].dt.year
check.value_counts()

#column named 'effective_to_month' which extracts month from 'effective_to_date'
marketing_customer_df_clean['effective_to_month'] = marketing_customer_df_clean['effective_to_date'].dt.month

#table with total complaints by policy type and month
total_complaints_by_policy_type_and_month = marketing_customer_df_clean.groupby(['policy_type','effective_to_month' ]).agg({'number_of_open_complaints':'sum'})
display(total_complaints_by_policy_type_and_month.head())

#test to change result into long format

test = total_complaints_by_policy_type_and_month.reset_index()
display(test.head())

pd.melt(test, id_vars = ['policy_type'], value_vars = ['effective_to_month','number_of_open_complaints'] )
```
